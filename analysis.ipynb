### 1. Set up environment

```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import yfinance as yf
from matplotlib import rcParams
rcParams['figure.figsize'] = 8, 6
import seaborn as sb
sb.set()
```

### 2. I will download the data for Amazon from `yfinance`

```python
amzn = yf.download('AMZN',  start = "2021-02-01" , end = "2023-02-04")
```

```python
amzn.head()
```

```python
amzn_close = amzn['Close']
amzn_return = pd.DataFrame(np.log(amzn_close).diff())
amzn_return.head()
```

### 3. After download and put the data into dataframe, I will store rate of return in some separate series, and plot the data
